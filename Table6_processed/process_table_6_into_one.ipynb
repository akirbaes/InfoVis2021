## In this notebook, the 4 seperate kingdom tables with the extinction details are merged into one

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### adding species name to tables

```python
animal=pd.read_csv('redlist/Table 6a Animal species (kingdom Animalia) by country show all.csv')
```

```python
animal.head()
```

```python
animal_df=pd.DataFrame(animal)
animal_df
```

```python
animal_df['species']='animal'
animal_df
```

```python
animal_df.to_csv('redlist/processedTable/Table6_Animals.csv')
```

```python
#adding species to Table6_plants
plant=pd.read_csv('redlist/Table 6b Plant species (kingdom Plantae) by country show all.csv')
plant
plant_df=pd.DataFrame(plant)
plant_df
plant_df['species']='plant'
plant_df
plant_df.to_csv('redlist/processedTable/Table6_Plants.csv')
```

```python
#adding species to Table6_chromist
chromist=pd.read_csv('redlist/Table 6d Chromist species (kingdom Chromista) by country - show all.csv')
chromist
chromist_df=pd.DataFrame(chromist)
chromist_df
chromist_df['species']='chromist'
chromist_df
chromist_df.to_csv('redlist/processedTable/Table6_Chromist.csv')
```

```python
#adding species to Table6_fungus
fungus=pd.read_csv('redlist/Table 6c Fungus species (kingdom Fungi) by country - show all.csv')
fungus
fungus_df=pd.DataFrame(fungus)
fungus_df
fungus_df['species']='fungus'
fungus_df
fungus_df.to_csv('redlist/processedTable/Table6_fungus.csv')
```

```python
#combining all 4 tables to single one Table6
# Table_7_merge(processed).csv
merge_table6=pd.concat([animal_df,plant_df,chromist_df,fungus_df],axis=0)
merge_table6.to_csv('redlist/processedTable/Table_6_merge(processed).csv')
merge_table6
```
